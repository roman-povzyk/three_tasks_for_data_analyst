## 1. Условие задачи

### 1.1. Вводные данные:

C целью определить лучший баннер в мобильном приложении, тестовой группе пользователей была предложена версия с другим цветовым решением с одинаковой возможность покупки платного функционала.
A/B-тест был проведен в период с 08.05.19 по 09.05.19. 

### 1.2. Задача:
Необходимо оценить какой из вариантов баннера является более предпочтительным для пользователей из региона US.

### 1.3. Данные:

Таблица Users

UserID –  идентификационный номер пользователей;
Country –  регион продажи;
Group – тип исследуемой группы.

Таблица Revenue

UserID –  идентификационный номер пользователей;
Date –  дата просмотра пользователем баннера;
Revenue – выручка от приобретения пользователем платного функционала.

## 2. Подготовка данных и решение

Для работы с данными воспользуемся программой Power BI.

Соединяем обе таблицы по UserID, оставляем в качестве страны только US и даты, когда проводился A/B-тест.

Разбиваем таблицу на две по целевой и контрольной группам.

Получаем данные о количестве подписчиков и прибыли от них

```python
# Прописываем полученые результаты
user_control = 3819
revenue_control = 217349.7
user_target = 4092
revenue_target = 253652.4
```

```python
# Вычисляем прибыль на пользователя с контрольной группы
revenue_per_user_control = revenue_control/user_control
round(revenue_per_user_control, 2)
```

```python
# Вычисляем прибыль на пользователя с целевой группы
user_per_user_target = revenue_target/user_target
round(user_per_user_target, 2)
```

```python
# Определям, на сколько процентов больше принес прибыли пользователь с целевой группы по сравнению с контрольной
round(((user_per_user_target - revenue_per_user_control) / revenue_per_user_control) * 100, 2)
```

## 3. Ответ
Таким образом, мы видим, что жителям США больше нравится новый банер по сравнению со старым. И он принес больше прибыли на пользователя за время AB-тестирования.
